```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "bat":["shahid","hamad","amir"],
    "size":[12,34,67]
}
df=pd.DataFrame(data)
plt.plot(df['bat'],df['size'])
plt.xlabel("bat (names)")
plt.ylabel("size (inches)")
plt.title("Number od scores for each batsman " ,loc='right')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(4)
b=[23,45,67,90]
c=[12,43,21,76]
fig=plt.figure()
ax=plt.subplot()
ax.plot(a,b,'k--',label="Frequency")
ax.plot(a,c,'ko:',label="Time")
ax.legend()
plt.title("Numbers of frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
students=np.array(["amna","ayesha","alia"])
marks=np.array([86,90,92])
plt.bar(students,marks)
plt.xlabel("students")
plt.ylabel("marks")
plt.title("childrean scores")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
cricketers=np.array(["aman","babr","kamran","afridi"])
score=np.array([23,89,546,90])
plt.pie(score,labels=cricketers , autopct='%1.2f%%')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
rain=np.array([12,45,12,89,45])
days=np.array(["monday","tuesday","wednesday","thursday","friday"])
plt.plot(days,rain)
plt.xlabel("rain")
plt.ylabel("days")
plt.show()
```
